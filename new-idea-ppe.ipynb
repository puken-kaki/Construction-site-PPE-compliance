```python
!pip install -U ultralytics

from ultralytics import YOLO

model = YOLO("yolo26s.pt")

results = model.train(
    data="/kaggle/working/dataset/PPE-CCTV-6/data.yaml",
    epochs=110,
    imgsz=640,
    batch=32,
    workers=4,
    cache=True,
    save=True,
    device=0,
    amp=True,
    rect=False,
    degrees=15.0,
    perspective=0.001,
    mosaic=1.0,
    mixup=0.1,
    name="ppe_model"
)
```
